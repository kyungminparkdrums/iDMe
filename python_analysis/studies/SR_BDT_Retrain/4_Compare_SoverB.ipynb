```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_legacy = pd.read_csv(f'../SR_OneBinBDT/csv/SR_Yields/SR_BDTv5_S_over_B.csv')
```

```python
df_new = pd.read_csv('./csv/SR_Yields/SR_S_over_B.csv')
```

## ctau = 1mm, delta = 0.1

```python
ctau = 1
delta = 0.1

m1s = [5, 10, 20, 30, 40, 50, 70, 90, 100]
```

```python
ratio = []

for m1 in m1s:
    df = df_new
    s_over_b_df = df[ (df.m1 == m1) & (df.delta == delta) & (df.ctau == ctau) ].iloc[:,-1].to_list()[0]
    s_over_b_legacy = df_legacy[ (df_legacy.m1 == m1) & (df_legacy.delta == delta) & (df_legacy.ctau == ctau) ].iloc[:,-1].to_list()[0]

    ratio.append(s_over_b_df/s_over_b_legacy)
```

```python
plt.plot(m1s, ratio, label='New (x-cleaning, good_vtx, BDT retrain)')

plt.legend()
plt.grid()

plt.xlabel('M1 [GeV]')
plt.ylabel('S/sqrt(B) Ratio to latest SR with legacy samples')

plt.ylim([0,4])

plt.title(f'ctau = {ctau}mm, delta = {delta}')
```
